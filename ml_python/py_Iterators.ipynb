# Iterator কী?

Iterator হলো এমন একটা object যেটা একের পর এক value return করে (sequence এর মতো), কিন্তু পুরো data একসাথে memory তে রাখে না।

```python
#example
L= [x for x in range(1, 10000)]

# for i in L:
  # print(i)

import sys
print(sys.getsizeof(L)/64)

x= range(1, 10000000)

# for i in x:
  # print(i)

print(sys.getsizeof(x)/64)

```

```python
# Python loop internally এটা করে:
for i in [1, 2, 3]:
  print(i)

#iternally

it= iter([1, 2, 3])

while True:
  try:
    x= next(it)
    print(x)
  except StopIteration:
    break
```

```python
#custhom iterators

class myIterator:
  def __init__(self, max):
    self.num= 1
    self.max= max

  def __iter__(self):
    return self

  def __next__(self):
    if self.num <= self.max:
      val= self.num
      self.num += 1
      return val
    else:
      raise StopIteration
```

```python
obj= myIterator(5)

for i in obj:
  print(i)
```

```python
def mera_khudka_for_loop(iterable):

    iterator = iter(iterable)

    while True:

        try:
            print(next(iterator))
        except StopIteration:
            break
```
