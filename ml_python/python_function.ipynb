```python
#create to function and passed two argument

def mySum(num1, num2):
  sum= num1 + num2
  print(f'Here is your sum number {sum}')


#function call with two values
mySum(5, 5)
```

```python
#create function

def mySum2(num3, num4):
  sum= num3 + num4
  return f'Here is to sum number: {sum}'

squares= mySum2(8,12)

print('Your Squares: ', squares)
```

```python
# function definition
def find_square(num):
    result = num * num
    return result

# function call
square = find_square(3)

print('Square:', square)
```
