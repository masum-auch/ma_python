```python
def return_sum(func, L):
  result = 0
  for i in L:
    if func(i):
      result = result + i

  return result
```

```python
L = [11, 14, 21, 23, 56, 78, 45, 29, 28]
```

```python
x = lambda x: x%2 == 0
y = lambda y: y%2 != 0
z = lambda z: z%3 ==0
```

```python
print(return_sum(x, L))
print(return_sum(y, L))
print(return_sum(y, L))
```

# Filter

```python
L2 = [1, 2, 3, 4, 5, 6, 7, 8]
```

```python
L2
```

```python
list(filter(lambda x: x>4, L2))
```

```python
fruits = ['orange', 'banana', 'apple', 'cherry', 'mango']
```

```python
filter(lambda fruit:'e' in fruit, fruits)
```

```python
list(filter(lambda fruit: 'e' in fruit, fruits))
```

# REDUCE Function

```python
import functools
```

```python
functools.reduce(lambda x, y: x+y, L4)
```

```python
functools.reduce(lambda x,y: x-y, L4)
```
